```python
df_ar = df.loc[lambda df: df['label'] == 'architecure']
df_ar = df_ar.groupby(df_ar.State)['label'].count().to_frame().rename(columns={"label": "architecure"})

df_tr = df.loc[lambda df: df['label'] == 'travel&adventure']
df_tr = df_tr.groupby(df_tr.State)['label'].count().to_frame().rename(columns={"label": "travel"})

df_fo = df.loc[lambda df: df['label'] == 'food&drinks']
df_fo = df_fo.groupby(df_fo.State)['label'].count().to_frame().rename(columns={"label": "food"})

df_art = df.loc[lambda df: df['label'] == 'art&culture']
df_art = df_art.groupby(df_art.State)['label'].count().to_frame().rename(columns={"label": "art"})

df_total = pd.DataFrame()
```

```python
df_total = df_ar
```

```python
df_total = df_total.merge(df_tr, left_on = 'State', right_on = 'State')
```

```python
df_total = df_total.merge(df_fo, left_on = 'State', right_on = 'State')
```

```python
df_total = df_total.merge(df_art, left_on = 'State', right_on = 'State')
```
